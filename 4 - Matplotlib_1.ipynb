# Matplotlip

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#grafikleri g√∂sterecek
%matplotlib inline
```

```python
yasListesi = [10,20,30,40,50,60,70,75]
kiloListesi = [20,30,80,85,86,70,70,75]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
# plot(x,y, "green")
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
```

```python
#etiket
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
#plt.show() demedik √ß√ºnk√º matplotlib inline kullanƒ±yorum
```

```python
#0 ile 10 arasƒ±nda 20 adet sayƒ±
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.xlabel("1. dizi")
plt.ylabel("2. dizi")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r.-")
plt.xlabel("1. dizi")
plt.ylabel("2. dizi")
```

# subplot (kac sƒ±ra,kac grafik,kacƒ±ncƒ±yƒ± ciz) 2 grafik yapma

```python
plt.subplot(1,2,1) # birinci grafik YANYANA
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.xlabel("1. dizi")
plt.ylabel("2. dizi")
plt.subplot(1,2,2) # ikinici grafik
plt.plot(numpyDizisi2,numpyDizisi1,"b.-")
plt.xlabel("2. dizi")
plt.ylabel("1. dizi")
```

## OBJE KULLANARAK YAPMA

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9]) # shift + tab
#                                grefaik konumu, width , height                                   
figureAxes.plot(numpyDizisi1,numpyDizisi2,"r")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("X ekseni veri ismi")
figureAxes.set_title("Grafik Ba≈ülƒ±ƒüƒ±")
plt.show()
```

## -----------

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
#label etiket verir
yeniEksen.plot(numpyDizisi1,numpyDizisi2, label = "numpyDizisi ** 2")
yeniEksen.legend()
```

```python
#2 adet veri
#figsize boyutu ayarladƒ±
#yeniFigur = plt.figure(figsize = (3,3))

#dpi boyutu
yeniFigur = plt.figure(dpi = 100)

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
#label etiket verir
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 **3, label = "numpyDizisi ** 3")
#loc pozisyon verir
yeniEksen.legend(loc=1)
```

```python
# grafigi kaydetme
yeniFigur.savefig("benimfigur.png",dpi=200)
```
